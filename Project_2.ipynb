### Text-based To-Do List Application

Below is a simple text-based To-Do list. You can run the code cells to interact with it. The tasks will be stored in a Python list.

**Features:**
- Add a new task
- View all tasks
- Mark a task as complete
- Delete a task

```python
tasks = []

def add_task(task_description):
    if task_description:
        tasks.append({'description': task_description, 'completed': False})
        print(f'Task "{task_description}" added.')
    else:
        print('Task description cannot be empty.')

def view_tasks():
    if not tasks:
        print('No tasks in the list.')
        return
    print('\n--- Your To-Do List ---')
    for i, task in enumerate(tasks):
        status = '✓' if task['completed'] else ' ' # Checkmark for completed tasks
        print(f'{i + 1}. [{status}] {task["description"]}')
    print('-----------------------\n')

def mark_task_complete(task_number):
    if 1 <= task_number <= len(tasks):
        tasks[task_number - 1]['completed'] = True
        print(f'Task {task_number} marked as complete.')
    else:
        print('Invalid task number.')

def delete_task(task_number):
    if 1 <= task_number <= len(tasks):
        task_description = tasks[task_number - 1]['description']
        del tasks[task_number - 1]
        print(f'Task "{task_description}" deleted.')
    else:
        print('Invalid task number.')

print("To-Do List functions initialized. Use add_task(), view_tasks(), mark_task_complete(), and delete_task() to manage your list.")
```

#### Example Usage:

```python
# Add some tasks
add_task("Buy groceries")
add_task("Finish report")
add_task("Call mom")

# View tasks
view_tasks()

# Mark a task as complete (e.g., the first task)
mark_task_complete(1)

# View tasks again to see the update
view_tasks()

# Add another task
add_task("Schedule dentist appointment")

# View tasks
view_tasks()

# Delete a task (e.g., the second task, which is "Finish report")
delete_task(2)

# View tasks one last time
view_tasks()
```
